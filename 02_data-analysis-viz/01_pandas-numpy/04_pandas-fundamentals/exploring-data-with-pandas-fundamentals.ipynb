<h2>02. Operaciones Vectorizadas</h2>
<br>
<div><p>Debido a que pandas está diseñado para funcionar como NumPy, se admiten muchos conceptos y métodos de Numpy. Recuerde que una de las formas en que NumPy facilita el trabajo con datos es con <strong>operaciones vectorizadas</strong>, u operaciones aplicadas a varios puntos de datos a la vez:</p>
<p><img src="https://s3.amazonaws.com/dq-content/289/vectorized.gif" alt="Operación vectorizada"></p>
<p>La vectorización no solo mejora el rendimiento de nuestro código, sino que también nos permite escribir código más rápidamente.</p>
<p>Debido a que pandas es una <em>extensión</em> de NumPy, también admite operaciones vectorizadas. Veamos un ejemplo de cómo funcionaría esto con una serie de pandas:</p>
</div>

```python

```



<ul>
<li><code>series_a + series_b</code> - Adición</li>
<li><code>series_a - series_b</code> - Resta</li>

<h2>05. Encadenamiento de métodos</h2>
<p>En el último ejercicio, usamos el método <code>Series.describe()</code> para explorar las columnas <code>rank</code> y <code>previous_rank</code>. Cuando revisó Es posible que haya notado algo extraño en los resultados: el valor mínimo para la columna <code>previous_rank</code> es <code>0</code>:</p>

```python
prev_rank = f500["previous_rank"]
print(prev_rank.describe())
```

```python
count    500.000000
mean     222.134000
std      146.941961
min        0.000000
25%       92.750000
50%      219.500000
75%      347.250000
max      500.000000
Name: previous_rank, dtype: float64
```
<p>Sin embargo, esta columna solo debe tener valores entre 1 y 500 (inclusive), por lo que un valor de <code>0</code> no tiene sentido. Para investigar la posible causa de este problema, confirmemos la número de valores <code>0</code> que aparecen en la columna <code>previous_rank</code>.</p>

<p>Recuerde que en la última lección, aprendimos a usar el método <code>Series.value_counts()</code> para mostrar los recuentos de los valores únicos en una columna:</p>

```python
countries = f500["country"]
countries_counts = countries.value_counts()
```

<p>En lugar de asignar la serie <code>countries</code> a su propia variable, podemos omitir ese paso y usar el método directamente en el resultado de la selección de la columna:</p>

```python
countries_counts = f500["country"].value_counts()
```

<p>Esto se llama <strong>encadenamiento de métodos</strong>, una forma de combinar varios métodos en una sola línea.</p>
<p>En la última lección, también aprendimos a usar <code>Series.loc[]</code> para seleccionar solo un elemento de una serie por etiqueta. Por ejemplo, para seleccionar solo los recuentos de <code>China</code>, usaríamos la siguiente línea de código:</p>

```python
print(f500["country"].value_counts().loc["China"])
```

```python
109
```

<p>A partir de aquí, verá el encadenamiento de métodos con más frecuencia. Al escribir código, siempre evalúe si el encadenamiento de métodos hará que su código sea más difícil de leer. Si es así, siempre es preferible dividir el código en más de una línea.</p>
<p>Usemos el método <code>Series.value_counts()</code> y <code>Series.loc</code> a continuación para confirmar el número de valores <code>0</code> en el <code columna >rank_anterior</code>.</p>

<p>En el último ejercicio, confirmamos que 33 empresas en el dataframe tienen un valor de <code>0</code> en la columna <code>previous_rank</code>. Dado que varias empresas tienen una clasificación <code>0</code>, podríamos concluir que estas empresas no tenían ninguna clasificación el año anterior. Al igual que usamos métodos de estadísticas descriptivas para explorar series individuales, también podemos usar métodos de estadísticas descriptivas para explorar nuestro dataframe <code>f500</code>. </p>
<p>Debido a que las series y los marcos de datos son dos objetos distintos, tienen sus propios métodos únicos. Sin embargo, hay muchas ocasiones en las que tanto los objetos de serie como los de dataframe tienen un método del mismo nombre que se comporta de manera similar. A continuación se muestran algunos ejemplos:</p>

<ul>
<li><a href="http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.max.html" target="_blank"><code>Series.max()</code></a> y <a href="http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.max.html" target="_blank "><code>dataframe.max()</code></a></li>
<li><a href="http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.min.html" target="_blank"><code>Series.min()</code></a> y <a href="http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.min.html" target="_blank "><code>dataframe.min()</code></a></li>
<li><a href="http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.mean.html" target="_blank"><code>Series.mean()</code></a> y <a href="http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.mean.html" target="_blank "><code>dataframe.media()</code></a></li>
<li><a href="http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.median.html" target="_blank"><code>Series.mediana()</code></a> y <a href="http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.median.html" target="_blank "><code>dataframe.median()</code></a></li>
<li><a href="http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.mode.html" target="_blank"><code>Series.mode()</code></a> y <a href="http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.mode.html" target="_blank "><code>dataframe.modo()</code></a></li>
<li><a href="http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.sum.html" target="_blank"><code>Series.sum()</code></a> y <a href="http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.sum.html" target="_blank "><code>dataframe.sum()</code></a></li></ul>

<p>A diferencia de sus contrapartes de serie, los métodos de dataframe requieren un <em>parámetro de eje</em> para que sepamos en qué eje calcular. Si bien puede usar números enteros para referirse al primer y segundo eje, los métodos de dataframe de pandas también aceptan las cadenas <code>index</code> y <code>columns</code> para el parámetro del eje :</p>
<p><img src="https://s3.amazonaws.com/dq-content/291/axis_param.svg" alt="parámetros del eje del dataframe"></p>
<p>Por ejemplo, si quisiéramos encontrar el valor mediano (medio) para las columnas <code>revenues</code> y <code>profits</code>, podríamos usar el siguiente código:</p>

```python
medians = f500[["revenues", "profits"]].median(axis=0)
# we could also use .median(axis="index")
print(medians)
```


<p>De hecho, el valor predeterminado para el parámetro del eje con estos métodos es <code>axis=0</code>. ¡Podríamos haber usado el método <code>median()</code> sin un parámetro para obtener el mismo resultado!</p>
