<h2>3. Variable Representation: Color</h2>
<br>
<div><p>We generated a scatter plot with two variables: <code>Gr Liv Area</code> and <code>SalePrice</code>. The scatter plot shows a clear positive correlation — the greater the living area, the bigger the sale price (typically, not always).</p>
<p></p><center>
<img src="screen2_2.png" alt="img">
</center><p></p>
<p>Notice that each dot (also called a marker) on the scatter plot is blue. We can change the color intensity of the dots to represent a new variable.</p>
<p>Below, we use the <code>hue</code> parameter to add the <code>Overall Qual</code> variable on the scatter plot. Recall that <code>Overall Qual</code> describes the quality ratings of the overall material and finish of the house.</p>
</div>

````python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice', hue='Overall Qual')
plt.show()
````

<p>The values in the <code>Overall Qual</code> variable range from one to ten — one is equivalent to "very poor," and ten is equivalent to "very excellent" (per the documentation).</p>
<p>Seaborn matched lower ratings with lighter colors and higher ratings with darker colors. A pale pink represents a rating of one, while black represents a ten. Seaborn also generated a legend to describe which color describes which rating.</p>

<p>Let's say we want the colors to vary between red and green — where dark red means a rating of one and dark green means a rating of ten. We can make this change using the <code>palette</code> parameter with the <code>'RdYlGn'</code> argument:</p>

````python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice', hue='Overall Qual', palette='RdYlGn')
plt.show()
````

<p>The argument <code>'RdYlGn'</code> contains three abbreviations:</p>

<ul>
<li><code>Rd</code>: red</li>
<li><code>Yl</code>: yellow</li>
<li><code>Gn</code>: green</li>
</ul>

<p>The argument <code>'RdYlGn'</code> as a whole describes a color palette that starts with red, goes through yellow, and ends with green. Below, we see a few color palettes:</p>

<center>
<img src="screen3_3.png" alt="img" width="600" height="400">
</center>

<p>We can see the color code to the left of each palette — this is the code that we can input into the <code>palette</code> parameter. The color palette we used for our scatter plot (<code>'RdYlGn'</code>) is the fifth one from the bottom.</p>

<p>We can find all the available color palettes in Matplotlib's <a href="https://matplotlib.org/tutorials/colors/colormaps.html#miscellaneous" target="_blank">documentation</a> (remember that Seaborn uses Matplotlib; when you click the documentation link, you'll need to scroll down to find the color palettes).</p>

<h2>4. Variable Representation: Size</h2>
<br>
<div><p>We've seen three ways of representing the values of a variable:</p>
<ul>
<li>We represented the values of <code>Gr Liv Area</code> as coordinates on the x-axis.</li>
<li>We represented the values of <code>SalePrice</code> as coordinates on the y-axis.</li>
<li>We represented the values of <code>Overall Qual</code> as colors.</li>
</ul>
<p>Another element we can use to represent values is size. A dot can have a color and x- and y-coordinates, but it can also be larger or smaller. Below, we use a size representation to add the <code>Garage Area</code> variable on the graph — we use the <code>size</code> parameter. Recall that <code>Garage Area</code> describes the garage area in square feet.</p>
</div>

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice',
            hue='Overall Qual', palette='RdYlGn',
            size='Garage Area')
plt.show()
```

<div>
<p></p><center>
<img src="screen4_1.png" alt="img">
</center><p></p>
<p>To make the size differences more visible, we'll increase the size range — the <code>sizes</code> parameter takes in a <code>tuple</code> specifying the minimum and maximum size.</p>    
</div>

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice',
            hue='Overall Qual', palette='RdYlGn',
            size='Garage Area', sizes=(1,300))
plt.show()
```

<div>
<p></p><center>
<img src="screen4_2.png" alt="img">
</center><p></p>
<p>The <code>sizes</code> parameter can take in a <code>tuple</code> only if the variable we represent is numerical — <code>Garage Area</code> is a numerical variable. The <code>tuple</code> in <code>sizes</code> represents a range. The minimum value in the range maps to the minimum value in the variable. Similarly, the maximum value in the range maps to the maximum value in the variable.</p>
<p>To control the sizes for a non-numerical (categorical) variable, we need to use a <code>list</code> or a <code>dict</code>. Instead of specifying the range, we need to specify the sizes for each unique value in the variable.</p>
<p>The <code>Rooms</code> variable is categorical, and it has two unique values: <code>'7 rooms or more'</code>, and <code>'6 rooms or less'</code>. Below, we pass in the list <code>[200,50]</code> to map <code>'7 rooms or more'</code> to a size of <code>200</code> and <code>'6 rooms or less'</code> to a size of <code>50</code>.</p>
</div>

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice',
            hue='Overall Qual', palette='RdYlGn',
            size='Rooms', sizes=[200,50])
plt.show()
```

<div>
<p></p><center>
<img src="screen4_3.png" alt="img">
</center><p></p>
<p>If we don't want to control the size, then Seaborn does the size mapping automatically anyway.</p></div>

<h2>5. Variable Representation: Shape</h2>
<br>
<p>Another visual property we can exploit is shape. On the graph we've built, each dot has the shape of a circle. Instead of a circle, however, it could have been a triangle, a square, etc.</p>

<p>More generally, we call the dots on our graphs <strong>markers</strong>. The marker can take various shapes: circle, triangle, square, etc.</p>

<p>Below, we add the <code>Rooms</code> variable by changing the shape of the markers. A circle now means a house with seven rooms or more, and an "x" sign represents a house with six rooms or less. To make this change, we use the <code>style</code> parameter.</p>

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice',
            hue='Overall Qual', palette='RdYlGn',
            size='Garage Area', sizes=(1,300),
            style='Rooms')
plt.show()
```

<div>
<p></p><center>
<img src="screen5_1.png" alt="img">
</center><p></p>
<p>By default, Seaborn chose a circle and an "x" sign to represent the values in the <code>Rooms</code> variable. Recall that <code>Rooms</code> has two unique values: <code>'7 rooms or more'</code>, and <code>'6 rooms or less'</code>.</p>
<p>If we're not happy with Seaborn's marker choice, we can change the markers' shape. To see the available shapes, we can check Matplotlib's <a href="https://matplotlib.org/3.3.3/api/markers_api.html" target="_blank">documentation</a>.</p>
<p>Below, we add different markers using the <code>markers</code> parameter. Each marker shape has a <code>string</code> representation that we can find in the documentation referenced earlier.</p>
</div>


```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice',
            hue='Overall Qual', palette='RdYlGn',
            size='Garage Area', sizes=(1,300),
            style='Rooms', markers=['*','v'])
plt.show()
```
<div>
<p></p><center>
<img src="screen5_2.png" alt="img">
</center><p></p>
<p>Choosing the best markers depends on each graph. For our situation, we need markers that clearly show the small differences in size and color. The star shapes above look nice, but the circles seem to offer a clearer representation in terms of size.</p>
</div>

<h2>6. Variable Representation: Spatial Separation</h2>
<br>

<div><p>So far, we've added the following variables on our graph:</p>
<ul>
<li><code>Gr Liv Area</code>: represented with x-coordinates.</li>
<li><code>SalePrice</code>: represented with y-coordinates.</li>
<li><code>Overal Qual</code>: represented with color.</li>
<li><code>Garage Area</code>: represented with size.</li>
<li><code>Rooms</code>: represented with shape.</li>
</ul>
<p>We'll add one more variable by spatially separating the graph based on the values in the <code>Year</code> variable. This variable describes the year when a house was built, and it has only two values: <code>1999 or older</code> and <code>2000 or newer</code>. For each value, we'll build a separate graph that will display the five variables we've already plotted.</p>
<p>Below, we add the <code>Year</code> column using the <code>col</code> parameter:</p>
</div>

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice',
            hue='Overall Qual', palette='RdYlGn',
            size='Garage Area', sizes=(1,300),
            style='Rooms', col='Year')
plt.show()
```
<div>
<p></p><center>
<img src="screen6_1.png" alt="img">
</center><p></p>
<p>Individually, each graph displays only five variables, but together they show six variables — the <code>Year</code> variable and the other five.</p>
<p>The quantity of information we see in the data visualization above is very large. Each column has 2,930 data points (exception: <code>Garage Area</code> has 2,929), which means we condensed 17,579 data points into one picture.</p>
<p>More importantly, however, we managed to visually represent the relationships between all six variables. Although there's so much information, the graph is readable and shows clear patterns.</p>
<p>The graph we built is essentially a scatter plot. However, because it shows the relationships between so many variables, we call it a <strong>relational plot</strong>.</p>
<p>We can do much more with Seaborn than just creating relational plots. We've shared a few Seaborn tutorials in the resources section of the takeaways to show you what this library can do.</p>
</div>
